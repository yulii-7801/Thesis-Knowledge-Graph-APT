```python
# 1. 直接从官方仓库重新下载原始的 run.py 覆盖掉被改坏的版本（秒级完成）
!wget -qO run.py https://raw.githubusercontent.com/malllabiisc/CompGCN/master/run.py

# 2. 执行严格匹配 Tab 缩进的替换脚本
python_code = """
with open("run.py", "r", encoding="utf-8") as f:
    lines = f.readlines()

with open("run.py", "w", encoding="utf-8") as f:
    for line in lines:
        if "sub, rel, obj = map(str.lower, line.strip().split('\\t'))" in line:
            indent = line[:len(line) - len(line.lstrip())] # 获取原代码的真实缩进（Tab）
            f.write(indent + "parts = line.strip().split()\\n")
            f.write(indent + "if len(parts) != 3:\\n")
            f.write(indent + "\\tcontinue\\n")  # 强制使用 Tab 缩进 continue
            f.write(indent + "sub, rel, obj = map(str.lower, parts)\\n")
        else:
            f.write(line)
print("容错逻辑注入成功，TabError 已消除！")
"""

with open("fix_indent.py", "w") as f:
    f.write(python_code)

!python fix_indent.py
```

```python
with open("run.py", "r", encoding="utf-8") as f:
    lines = f.readlines()

modified = False
with open("run.py", "w", encoding="utf-8") as f:
    for line in lines:
        # 只抓取最核心、最不容易因为转义符写错的半句话
        if "map(str.lower, line.strip().split(" in line:
            indent = line[:len(line) - len(line.lstrip())] # 自动获取原代码的 Tab 缩进
            f.write(indent + "parts = line.strip().split('\\t')\n")
            f.write(indent + "if len(parts) != 3:\n")
            f.write(indent + "\tcontinue\n") # 强制用 Tab 缩进 continue
            f.write(indent + "sub, rel, obj = map(str.lower, parts)\n")
            modified = True
        else:
            f.write(line)

if modified:
    print("【真正成功】空行免疫护盾已切实写入 run.py！代码已被修改。")
else:
    print("【失败】未找到目标代码，请确保你的 run.py 是官方原始版本。")
```
